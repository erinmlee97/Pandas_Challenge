```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
#Find Total Players
total_players = len(purchase_data["SN"].unique())
total_df = pd.DataFrame({"Total Players": [total_players]})
total_df
```

```python
purchase_data['Price'] = pd.to_numeric(purchase_data['Price'])
```

```python
# Find the following:
# Number of unique items
unique_items = len(purchase_data["Item Name"].unique())

# AveragePrice
avg_price = purchase_data["Price"].mean()

# Number of Purchases
number_purchases = len(purchase_data["Item Name"])

# Total Revenue 
total_revenue = purchase_data["Price"].sum()

# Create a new DataFrame
purchasing_analysis = pd.DataFrame({"Number of Unique Items": [unique_items],
                                  "Average Price": [avg_price], 
                                  "Number Of Purchases": [number_purchases],
                                  "Total Revenue": [total_revenue]})  
#Format DataFrame
purchasing_analysis["Average Price"] = purchasing_analysis["Average Price"].map("${0:,.2f}".format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].map("${0:,.2f}".format)
                                                                                
purchasing_analysis
```

```python
#Gender Demographics

#Create dataFrame for Gender
gender = purchase_data[["SN", "Gender"]]

#Drop duplicate SN to not count the same person twice
gender = gender.drop_duplicates()
count = gender["Gender"].value_counts()

#Get Counts for Genders
total_counts = [count[0],count[1], count[2]]
percents = [round((count[0]/total_players)*100,2),round((count[1]/total_players)*100,2),round((count[2]/total_players)*100,2)]

#Get Percentages for Genders

#Create DataFrame
gender_df = pd.DataFrame({"Total Counts": total_counts,
                          "Percentage of Players": percents})
gender_df.index = (["Male", "Female", "Other/ Non-Disclosed"])

#Format DataFrame
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map("{:.2f}%".format)
gender_df
```

```python
#Purchasing Analysis by Gender

#Find the Purchase Count
purchase_counts = purchase_data.groupby(["Gender"]).count()["Price"]

#Find average purchase price
avg_purchase_price=purchase_data.groupby(["Gender"]).mean()["Price"]

#Find total purchase value
total_purchase_value=purchase_data.groupby(["Gender"]).sum()["Price"]

#Find average total purchase per person
avg_total_person = total_purchase_value/count

#Create DataFrame
analysis_by_gender = pd.DataFrame({"Purchase Count": purchase_counts,
                                  "Average Purchase Price": avg_purchase_price,
                                  "Total Purchase Value": total_purchase_value,
                                  "Avg Total Purchase per Person": avg_total_person})

#Format DataFrame
analysis_by_gender["Average Purchase Price"] = analysis_by_gender["Average Purchase Price"].map("${0:,.2f}".format)
analysis_by_gender["Total Purchase Value"] = analysis_by_gender["Total Purchase Value"].map("${0:,.2f}".format)
analysis_by_gender["Avg Total Purchase per Person"] = analysis_by_gender["Avg Total Purchase per Person"].map("${0:,.2f}".format)

analysis_by_gender
```

```python
#Age Demographics

#Create Bins
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,999999]
group_names = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

#Sort values into bins
purchase_data["Age Range"] = pd.cut(purchase_data["Age"],bins, labels=group_names)

#Create new DataFrame and group by age
age_ranges = purchase_data.groupby("Age Range")

#Find total players by age
age_demo_total = age_ranges["SN"].nunique()

#Find percent by age
age_demo_percent = (age_demo_total/total_players) * 100

#Create a new DataFrame
age_demo_df = pd.DataFrame({"Total Count": age_demo_total, 
                           "Percentage of Players": age_demo_percent})
#Format DataFrame
age_demo_df["Percentage of Players"] = age_demo_df["Percentage of Players"].map("{:.2f}%".format)

age_demo_df
```

```python
#Purchase Analysis by Age

#Find the purchase count by age group
purchase_count_age = age_ranges["Purchase ID"].count()

#Find the average purchase prince by age group
avg_age_price = age_ranges["Price"].mean()

#Find total purchase value by age group
value_age = age_ranges["Price"].sum()

#Find the average purchase price per person in each age group
avg_per_person = value_age/age_demo_total

#Create DataFrame
analysis_age = pd.DataFrame({"Purchase Count": purchase_count_age, 
                             "Average Purchase Price": avg_age_price, 
                            "Total Purchase Value":value_age,
                            "Average Total Purchase per Person": avg_per_person})

#Format DataFrame
analysis_age["Average Purchase Price"] = analysis_age["Average Purchase Price"].map("${0:,.2f}".format)
analysis_age["Total Purchase Value"] = analysis_age["Total Purchase Value"].map("${0:,.2f}".format)
analysis_age["Average Total Purchase per Person"] = analysis_age["Average Total Purchase per Person"].map("${0:,.2f}".format)

analysis_age
```

```python
#Top Spenders

#Group spenders by SN
spender = purchase_data.groupby("SN")

#Find total purchases per spender
spender_count = spender["Purchase ID"].count()

#Find the average purchase price by spender
spender_avg = spender["Price"].mean()

#Find the total purchase value per spender
spender_total = spender["Price"].sum()

#Create new DataFrame
top_spender = pd.DataFrame({"Purchase Count": spender_count,
                           "Average Purchase Price": spender_avg,
                           "Total Purchase Value": spender_total})

#Format so that top spenders are on top
top_spender = top_spender.sort_values(["Total Purchase Value"], ascending=False)

#Format DataFrame
top_spender["Average Purchase Price"] = top_spender["Average Purchase Price"].map("${0:,.2f}".format)
top_spender["Total Purchase Value"] = top_spender["Total Purchase Value"].map("${0:,.2f}".format)

top_spender.head()
```

```python
#Most Popular Items

#Create new DtatFrame with information we need
items =purchase_data[["Item ID", "Item Name", "Price"]]

#Group by the items
item_stats = items.groupby(["Item ID", "Item Name"])

#Find how many times an item has been purchased
item_count = item_stats["Price"].count()

#Find the purchase value per item
item_value = item_stats["Price"].sum()

#Find the individual item price
item_price = item_value/item_count

#Create DataFrame
popular_items = pd.DataFrame({"Purchase Count": item_count, 
                              "Item Price": item_price,
                             "Total Purchase Value": item_value})

#Sort DataFrame
popular_items = popular_items.sort_values(["Purchase Count"], ascending=False)

#Format DataFrame
popular_items["Item Price"] = popular_items["Item Price"].map("${0:,.2f}".format)
popular_items["Total Purchase Value"] = popular_items["Total Purchase Value"].map("${0:,.2f}".format)

popular_items.head()
```

```python
#Most Profitable Items

#Create new DataFrame 
profitable_items= pd.DataFrame({"Purchase Count": item_count, 
                              "Item Price": item_price,
                             "Total Purchase Value": item_value})

#Sort DataFrame
profitable_items = profitable_items.sort_values(["Total Purchase Value"], ascending=False)

#Format DataFrame
profitable_items["Item Price"] = profitable_items["Item Price"].map("${0:,.2f}".format)
profitable_items["Total Purchase Value"] = profitable_items["Total Purchase Value"].map("${0:,.2f}".format)

profitable_items.head()
```
